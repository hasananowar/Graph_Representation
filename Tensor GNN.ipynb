```python
import torch

# Number of nodes in the graph
num_nodes = 3

# Number of time steps
num_time_steps = 2

# Temporal edge information (random weights for illustration)
temporal_edge_weights = torch.rand(num_nodes, num_nodes, num_time_steps)

# Print the temporal edge weights tensor
print("Temporal Edge Weights Tensor:")
print(temporal_edge_weights)

print(temporal_edge_weights.shape)
```

```python
import torch
import torch.nn as nn

# Number of nodes in the graph
num_nodes = 5

# Number of time steps
num_time_steps = 3

# Temporal edge information (random weights for illustration)
temporal_edge_weights = torch.rand(num_nodes, num_nodes, num_time_steps)

# Define an embedding layer
embedding_dim = 16
embedding_layer = nn.Embedding(num_nodes, embedding_dim)

# Flatten the temporal edge weights tensor for embedding
flat_temporal_edge_weights = temporal_edge_weights.view(-1, num_nodes)

# Apply the embedding layer
embedded_temporal_edge_weights = embedding_layer(flat_temporal_edge_weights.long())

# Reshape back to the original dimensions
embedded_temporal_edge_weights = embedded_temporal_edge_weights.view(num_nodes, num_nodes, num_time_steps, embedding_dim)
# print("weights")
# print(temporal_edge_weights)

# Print the embedded tensor
print("Embedded Temporal Edge Weights Tensor:")
print(embedded_temporal_edge_weights)

print(embedded_temporal_edge_weights.shape)
```

```python
# Load nodes from a Pandas DataFrame
# Example DataFrame with columns X, Y, Z, Atom_Name, Residue_Name, Residue_ID, Atom_Type, and Timeframe
# Load nodes for all timesteps from a .npy file
filepath = 'continuous_edge.csv'  # Replace with the actual file path
df = pd.read_csv(filepath)
df
```

```python
# Extract 'Node1' and 'Node2' columns
nodes_data = df[['Node1', 'Node2']]

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Fit and transform 'Node1' and 'Node2'
nodes_encoded = encoder.fit_transform(nodes_data)

# Create new column names for the one-hot encoded features
encoded_columns = encoder.get_feature_names_out(['Node1', 'Node2'])

# Replace the original 'Node1' and 'Node2' columns with the one-hot encoded features
df_encoded = pd.concat([df.drop(['Node1', 'Node2'], axis=1), pd.DataFrame(nodes_encoded, columns=encoded_columns)], axis=1)

# Print the encoded DataFrame
df_encoded
```

```python
# Assuming 'filtered_edge_info' is a DataFrame with 'Edge_Type' column
edge_type = df['Edge_Type'].values.astype(int)

# Edge type features using PyTorch
num_edge_types = len(set(edge_type))
print("Number of Edge Types:", num_edge_types)

# One-hot encode the edge types directly using PyTorch
edge_type_feats = torch.nn.functional.one_hot(torch.from_numpy(edge_type-1), num_classes=num_edge_types)

# Print the one-hot encoded features and their dimensions
print("One-Hot Encoded Features:")
print(edge_type_feats)
print("Feature Dimensions:", edge_type_feats.size(1))
```
